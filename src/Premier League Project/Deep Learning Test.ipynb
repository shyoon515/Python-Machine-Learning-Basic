```python
from sklearn.model_selection import train_test_split

# 데이터를 DataFrame 형태로 불러 옵니다.
ml_data = pd.read_csv('transformed_data.csv', encoding='cp949')
ml_data = ml_data[:-3]

# 의미없는 변수는 삭제합니다.
ml_data = ml_data.drop(columns=['Season', 'HomeTeam', 'AwayTeam', 'LSR', 'CSR'])

X = ml_data.drop(columns=['Result'])
y = ml_data['Result'].astype('int32')
y = y.replace({3:2, 1:1, 0:0})

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# 학습용 데이터를 tf.data.Dataset 형태로 변환합니다. from_tensor_slices 함수를 사용하여 변환하고 batch를 수행하게 합니다.
train_ds = tf.data.Dataset.from_tensor_slices((X_train.values, y_train.values))
train_ds = train_ds.shuffle(len(X_train)).batch(batch_size=5)

"""
1. keras를 활용하여 신경망 모델을 생성합니다.
   3가지 범주를 갖는 label 데이터를 분류하기 위해서 마지막 레이어 노드를 아래와 같이 설정합니다.
"""
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(9, input_dim=14),
    tf.keras.layers.Dense(8, activation='sigmoid'),
    tf.keras.layers.Dense(5, activation='sigmoid'),
    tf.keras.layers.Dense(3, activation='softmax')
    ])
#
# 학습용 데이터를 바탕으로 모델의 학습을 수행합니다.
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(train_ds, epochs=30, verbose=2)

# 테스트용 데이터를 바탕으로 학습된 모델을 평가합니다.
loss, acc = model.evaluate(X_test, y_test)

# 테스트용 데이터의 예측값을 구합니다.
predictions = model.predict(X_test)

# 결과를 출력합니다.
print("테스트 데이터의 Accuracy 값: ", acc)
for i in range(5):
    print("%d 번째 테스트 데이터의 실제값: %d" % (i, y_test.iloc[i]))
    print("%d 번째 테스트 데이터의 예측값: %d" % (i, np.argmax(predictions[i])))
```
